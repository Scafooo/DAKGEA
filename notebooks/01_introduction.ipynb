# DAKGEA - Introduzione

Questo notebook introduce DAKGEA (Data Augmentation for Knowledge Graph Entity Alignment) e mostra come utilizzare il framework per migliorare l'allineamento di entità tra knowledge graph.

## Cos'è DAKGEA?

DAKGEA è un framework per:
- **Data Reduction**: Ridurre il dataset mantenendo le entità più rappresentative
- **Data Augmentation**: Generare dati sintetici usando modelli di linguaggio (PLM)
- **Entity Alignment**: Allineare entità tra knowledge graph diversi
- **Evaluation**: Valutare la qualità dell'allineamento

## Pipeline

```
Dataset Originale
       ↓
  Reduction (es. 30% delle entità)
       ↓
  Augmentation (genera dati sintetici)
       ↓
  Training + Evaluation
       ↓
    Risultati
```
