# DAKGEA - Introduction

This notebook introduces DAKGEA (Data Augmentation for Knowledge Graph Entity Alignment) and shows how to use the framework to improve entity alignment between knowledge graphs.

## What is DAKGEA?

DAKGEA is a framework for:
- **Data Reduction**: Reduce the dataset while keeping the most representative entities
- **Data Augmentation**: Generate synthetic data using Pre-trained Language Models (PLM)
- **Entity Alignment**: Align entities between different knowledge graphs
- **Evaluation**: Evaluate the quality of the alignment

## Pipeline

```
Original Dataset
       ↓
  Reduction (e.g., 30% of entities)
       ↓
  Augmentation (generate synthetic data)
       ↓
  Training + Evaluation
       ↓
    Results
```
